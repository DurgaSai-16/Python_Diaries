###  Variables

* Memory location to store a value (Value can be string, numeric, etc)
* Variable is assigned to value before referencing and it can accessed, updated later.
* Rules:
    - Should begin with a letter or underscore.
    - Should never begin with a number.
    - Should never be a reserved word.
    - Case sensitive.
    - Tip: Short and cute names will be better
* Data type of a variable can be known with the `type()` function.
* Casting can be done to specify the data type of a variable.


````python
Name = "Vikram" #String
Jersey = 7 #Integer
BMI = 18.5 #Float
print(Name)
print(Jersey)
print(BMI)
print(type(Name)) #Type
print(int(BMI))  #Casting
````

### Data Types

* Classification of data items and represents various operation that can be performed on specific data.
* Standard data types in python are:
    1. Numeric
        1. Integer
        2. Complex number
        3. Float
    2. Boolean
    3. Sets
    4. Dictionary
    5. Sequence type
        1. Strings
        2. Lists
        3. Tuples

#### Numeric Data Types

````python
a = 9    # int
b = 1.6  # float
c = 7j   # complex
print(type(a))
print(type(b))
print(type(c))
#Type conversion below:
print(complex(b))
print(int(b))
print(float(a))
````

#### String Data Types

````python

Company = "TATA" #String
Employee = "RAGHUVARAN"
Stream = "BTech"
print(type(a)) #type casting
print(Company[0]) #Indexing
print(Employee[1:3])
print(Employee[5:]) #If Ending Index is not mentioned it prints till the end.
print(Employee[:]) #If start and end indices are not mentioned, it prints whole string
print(Employee[-1]) #-1 means Last letter
print(Stream.upper()) #Uppercase
print(Stream.lower()) #Lowercase
print(Employee + Stream) #Joining --String Concatenation
print(len(Employee)) #Length
print(Stream.replace('B','M')) #Replacing
print(Company[::-1]) #Reverse a string -- String slicing
print(Employee.split('U'))

````
